```python
import pandas as pd
import matplotlib.pyplot as plt

# Load a sample dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
data = pd.read_csv(url)

# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(data.head())

# Basic statistical summary
print("\nStatistical summary of the dataset:")
print(data.describe())

# Check for missing values
print("\nMissing values in the dataset:")
print(data.isnull().sum())

# Plotting
plt.figure(figsize=(10, 6))

# Histogram of the 'sepal_length' column
plt.subplot(2, 2, 1)
plt.hist(data['sepal_length'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')

# Scatter plot of 'sepal_length' vs 'sepal_width'
plt.subplot(2, 2, 2)
plt.scatter(data['sepal_length'], data['sepal_width'], color='green', edgecolor='black')
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

# Box plot of the 'petal_length' column
plt.subplot(2, 2, 3)
plt.boxplot(data['petal_length'], patch_artist=True, boxprops=dict(facecolor='lightblue', color='black'))
plt.title('Box Plot of Petal Length')
plt.ylabel('Petal Length')

# Pair plot to visualize relationships between variables
plt.subplot(2, 2, 4)
pd.plotting.scatter_matrix(data, alpha=0.2, figsize=(10, 10), diagonal='kde')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```
