```python
# Step 1: Import necessary libraries
import pandas as pd
import numpy as np  # Import NumPy correctly
import matplotlib.pyplot as plt

# Step 2: Load data
# For demonstration, let's create a DataFrame manually. In practice, you might load this from a csv file.
data = {
    'Date': pd.date_range(start='1/1/2020', periods=100, freq='D'),
    'Sales': np.random.randint(100, 500, size=(100)),  # Use np.random.randint
    'Category': np.random.choice(['Electronics', 'Clothing', 'Food'], size=(100))  # Use np.random.choice
}
df = pd.DataFrame(data)

# Step 3: Display the first few rows of the dataframe
print("First few rows of the dataset:")
print(df.head())

# Step 4: Basic data manipulations
# Group data by 'Category' and calculate total sales per category
total_sales_by_category = df.groupby('Category')['Sales']. sum()
print("\nTotal sales by category:")
print(total_sales_by_category)

# Step 5: Create a plot
# Let's plot the sales over time for each category
plt.figure(figsize=(10, 5))
for category in df['Category'].unique():
    subset = df[df['Category'] == category]
    plt.plot(subset['Date'], subset['Sales'], label=category)

plt.title('Sales Over Time by Category')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()

```
